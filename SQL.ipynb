
#### <span style="color: blue">Method - 2</span> &nbsp;&nbsp;<font size="+1">=====> &nbsp;&nbsp;using &nbsp;<code style="background-color: yellow">SQLALchemy engine</code> </font>


```python

from sqlalchemy import create_engine

# Define connection parameters
username = 'root'            # username
password = 'Mishra@1992'     # password
host = '127.0.0.1'           # or 'localhost'
port = '3306'                # default MySQL port
database = 'practice_db'     # your databse name

# Construct the connection string
connection_string = f'mysql://{username}:{password}@{host}:{port}/{database}'

# Create SQLAlchemy engine
engine = create_engine(connection_string)
```

```python

# Test the connection by executing a sample query
with engine.connect() as connection:
    result = connection.execute('SELECT 1')
    print(result.fetchone())
    
```


#### <span style="color: blue">Method - 3</span> &nbsp;&nbsp;<font size="+1">=====> &nbsp;&nbsp;using &nbsp;<code style="background-color: yellow">MySQLdb</code>&nbsp; and &nbsp;<code style="background-color: yellow">pandas.read_sql</code>&nbsp; </font>


```python
import MySQLdb

# Define connection parameters
host = '127.0.0.1'           # hostname or your MySQL server IP address
username = 'root'            # username
password = 'Mishra@1992'     # password
database = 'practice_db'     # database

try:
    # Establish connection
    connection = MySQLdb.connect(host = host,
                                 user = username,
                                 password = password,
                                 database = database)
    
    # Create a cursor object
    cursor = connection.cursor()     # <mysql.connector.connection_cext.CMySQLConnection object at 0x000001E9C5EF8610>

    # Execute a SQL query
    query = "SELECT * FROM employees"
    cursor.execute(query)

    # Fetch results
    results = cursor.fetchall()

    # Do something with the results
    for row in results:
        print(row)

    # Don't forget to close the cursor and connection when done
    cursor.close()
    connection.close()

except MySQLdb.Error as e:
    print("Error connecting to MySQL:", e)
```

<span style="color: red">To handle connection errors, use the **try** statement and **catch** all errors using the <code>errors.Error</code> exception:</span>

```python

import mysql.connector
from mysql.connector import errorcode


try:
  db = mysql.connector.connect(host = "127.0.0.1",
                                user = "root",
                                password = 'Mishra@1992',
                                database = 'practice_db')
except mysql.connector.Error as err:
  if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
    print("Something is wrong with your user name or password")
  elif err.errno == errorcode.ER_BAD_DB_ERROR:
    print("Database does not exist")
  else:
    print(err)
# else:
#   db.close()
    
```

```python

import mysql.connector
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

db = mysql.connector.connect(
    host = "127.0.0.1",
    user = "root",
    password = 'Mishra@1992',
    database = 'practice_db'
)

query = '''\
with cte as (
	select 
		*,
		lag(free, 1) over (order by seat_id) as previous_free,
		lead(free, 1) over (order by seat_id) as next_free
	from cinema
)
select * from cte
where free = 1 and (free = previous_free or free = next_free)
'''

pd.read_sql(query, db)
```
